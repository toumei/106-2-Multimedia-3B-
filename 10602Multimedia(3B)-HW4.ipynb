```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np

img = data.chelsea()
img2 = img.copy()

plt.figure(num = 'chelsea', figsize = (8, 8))
plt.subplot(1,2,1)
plt.title('origin img')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('modified img')
for i in range(0,50):
    for j in range(0,50):
        img2[i,j,1] = 0
plt.imshow(img2)

plt.show()
```

```python
from skimage import io, data, img_as_float, img_as_ubyte

img = io.imread('lena_std.bmp')
rows, cols, dims = img.shape

img2 = img.copy()
img3 = img.copy()
img4 = img.copy()

img2 = img_as_float(img2)
for i in range(0,rows):
    for j in range(0,cols):
        for k in range(0,3):
            img2[i,j,k] = 1.0 - img2[i,j,k]

img3 = img_as_ubyte(img3)
for i in range(0,rows):
    for j in range(0,cols):
        for k in range(0,3):
            img3[i,j,k] = 255 - img3[i,j,k]

img4 = 1 - img4
io.imshow(img4)
io.show()
```

```python
from skimage import io, data

img = io.imread('lena_std.bmp')
rows, cols, dims = img.shape
img2 = img[:,:,0]

for i in range(0,rows):
    for j in range(0,cols):
        sum = 0.0
        for k in range(0,3):
            sum = sum + img[i,j,k]
        img2[i,j] = sum / 3

io.imshow(img2)
io.show()

io.imsave('lena_std3.bmp', img2)
```

```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np

img = io.imread('lena_std.bmp')
rows, cols, dims = img.shape

img2 = np.uint8(np.zeros((rows,cols,3)))
plt.figure(figsize = (8,8))
plt.subplot(1,2,1)
plt.title('origin img')
plt.imshow(img)

plt.subplot(1,2,2)
plt.title('grey img')
for i in range(0,rows):
    for j in range(0,cols):
        sum = 0.0
        for k in range(0,3):
            sum = sum + img[i,j,k]
        for k in range(0,3):
            img2[i,j,k] = np.uint8(sum / 3)
plt.imshow(img2)

plt.show()
```

```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np

img = io.imread('lena_std.bmp')
rows, cols, dims = img.shape
row2 = int(rows/2)
col2 = int(cols/2)

img2 = img

for i in range(0,row2):
    for j in range(0,cols):
        for k in range(0,3):
            img2[i,j,k] = img[i,j,k]
            
for i in range(row2,rows):
    for j in range(0,cols):
        for k in range(0,3):
            img2[i,j,k] = img[row2+(-1)*(i-row2),j,k]
plt.imshow(img2)
plt.show()
```
