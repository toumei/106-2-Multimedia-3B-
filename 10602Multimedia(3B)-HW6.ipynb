```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np

img = data.chelsea()
rows, cols, dims = img.shape
img2 = img[:,:,0].copy()

threshold = 128
for i in range(0,rows):
    for j in range(0,cols):
        if img2[i,j] > threshold:
            img2[i,j] = 255
        else:
            img2[i,j] = 0
            
print(img2.shape)
io.imshow(img2)
io.show()
```

```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np

img = data.chelsea()
rows, cols, dims = img.shape
threshold = 128

img2 = np.uint8(np.zeros((rows,cols,3)))
plt.figure(figsize = (8,8))
plt.subplot(1,2,1)
plt.title('origin img')
plt.imshow(img)

for i in range(0,rows):
    for j in range(0,cols):
        for k in range(0,3):
            if img[i,j,0] > threshold:
                img2[i,j,k] = 255
            else:
                img2[i,j,k] = 0            
print(img2.shape)
plt.subplot(1,2,2)
plt.title('grey img')
plt.imshow(img2)
io.imshow(img2)
```

```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np

img = data.chelsea()
rows, cols, dims = img.shape
io.imshow(img)
half = np.matrix([[0,128],[192,64]])
img2 = img[:,:,0].copy()

for i in range(0,rows):
    for j in range(0,cols):
        index_i = i%2
        index_j = j%2
        if img2[i,j] > half[index_i,index_j]:
            img2[i,j] = 255
        else:
            img2[i,j] = 0 
plt.figure()
io.imshow(img2)
io.show()
```
