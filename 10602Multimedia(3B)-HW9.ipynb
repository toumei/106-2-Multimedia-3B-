```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np
import math

def posterize(img):
    rows, cols, dims = img.shape
    for i in range(0,rows):
        for j in range(0,cols):
            for k in range(0,3):
                clr = img[i,j,k]
                img[i,j,k] = math.floor(clr/64)*64+31
    return img

img = io.imread('scene.jpg')
io.imshow(img)

img2 = posterize(img)
plt.figure()
io.imshow(img2)
io.show()
```

```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np
import math

def halfBringhter(img):
    rows, cols, dims = img.shape
    ret_img = np.uint8(np.zeros((rows,cols,3)))
    for i in range(0,rows):
        for j in range(0,cols):
            for k in range(0,3):
                if j>cols/2:
                    ret_img[i,j,k] = np.uint8(img[i,j,k]/2)+128
                else:
                    ret_img[i,j,k] = img[i,j,k]
    return ret_img

img = io.imread('scene.jpg')
io.imshow(img)

img2 = halfBringhter(img)
plt.figure()
io.imshow(img2)
io.show()
```

```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np
import math

def sepiaToned(img):
    rows, cols, dims = img.shape
    for i in range(0,rows):
        for j in range(0,cols):
            red = img[i,j,0]
            blue = img[i,j,2]
            if red<63:
                red = red*1.1
                blue = blue*0.9
            if red>62 and red<192:
                red = red*1.15
                blue = blue*0.85
            if red>191:
                red = red*1.08
            if red>255:
                red = 255
            blue = blue*0.93
            
            img[i,j,0] = red
            img[i,j,2] = blue
    return img

img = io.imread('road.jpg')
plt.figure(num = 'road', figsize = (15,15))
plt.subplot(2,2,1)
plt.title('origin img')
plt.imshow(img)

img2 = sepiaToned(img.copy())
plt.subplot(2,2,2)
plt.title('sepia toned')
plt.imshow(img2)

img3 = sepiaToned(img2.copy())
plt.subplot(2,2,3)
plt.title('sepia toned2')
plt.imshow(img3)

img4 = sepiaToned(img3.copy())
plt.subplot(2,2,4)
plt.title('sepia toned3')
plt.imshow(img4)

plt.show()
```

```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np
import math

def edgeDetection(img):
    rows, cols, dims = img.shape
    for i in range(0,rows-1):
        for j in range(0,cols-1):
            sum = 0
            sum_rightBottom = 0
            for k in range(0,3):
                sum = sum + img[i,j,k]
                sum_rightBottom = sum_rightBottom + img[i+1,j+1,k]
            dif = abs(sum_rightBottom - sum)
            
            for k in range(0,3):
                img[i,j,k] = dif
    return img

img = io.imread('road.jpg')
plt.figure(num = 'road', figsize = (20,20))
plt.subplot(1,2,1)
plt.title('origin img')
plt.imshow(img)

img2 = edgeDetection(img)
plt.subplot(1,2,2)
plt.title('edge detection')
plt.imshow(img2)

plt.show()
```

```python
from skimage import io, data
import matplotlib.pyplot as plt
import numpy as np
import math

def edgeDetection(img, threshold):
    rows, cols, dims = img.shape
    for i in range(0,rows-1):
        for j in range(0,cols-1):
            sum = 0
            sum_rightBottom = 0
            for k in range(0,3):
                sum = sum + img[i,j,k]
                sum_rightBottom = sum_rightBottom + img[i+1,j+1,k]
            dif = abs(sum_rightBottom - sum)
            if dif > threshold:
                value = 0
            else:
                value = 255
            for k in range(0,3):
                img[i,j,k] = value
    return img

img = io.imread('road.jpg')
plt.figure(num = 'road', figsize = (20,20))
plt.subplot(1,2,1)
plt.title('origin img')
plt.imshow(img)

img2 = edgeDetection(img, 50)
plt.subplot(1,2,2)
plt.title('edge detection')
plt.imshow(img2)

plt.show()
```
